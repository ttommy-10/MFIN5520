```python
import math
import pandas as pd
import numpy as np
from scipy.stats import norm
```

```python
stockprice = 19456.71
t = 3
rfr = .0261
dividendyield = .0255
vol = .2218
sims = 10000
```

```python
dffinal = pd.DataFrame(columns=['trial','randomshock','stock','antithetic','return','antitheticreturn'])
for i in range(0,sims):
    randshock = np.random.uniform(low=0, high=1)
    stockp = stockprice*math.exp(((rfr-dividendyield-(vol**2)/2)*t)+vol*norm.ppf(randshock)*math.sqrt(t))
    antistockp = stockprice*math.exp(((rfr-dividendyield-(vol**2)/2)*t)-vol*norm.ppf(randshock)*math.sqrt(t))
    dffinal = dffinal.append({'trial':i+1,
                    'randomshock':randshock,
                    'stock':stockp,
                    'antithetic':antistockp,
                    'return': np.log(stockp/stockprice),
                    'antitheticreturn': np.log(antistockp/stockprice)},ignore_index=True)

```

```python
dffinal1=dffinal.copy()
dffinal1['pvpayoff']=np.where(dffinal1['return']<=0,1000*math.exp(-rfr*t),(1000+1000*dffinal['return']*0.2)*math.exp(-rfr*t))
dffinal1['pvantitheticpayoff']=np.where(dffinal1['antitheticreturn']<=0,1000*math.exp(-rfr*t),(1000+1000*dffinal['antitheticreturn']*0.2)*math.exp(-rfr*t))
dffinal1['Averagepayoff']= (np.mean(dffinal1['pvpayoff']) + np.mean(dffinal1['pvantitheticpayoff']))/2
```
